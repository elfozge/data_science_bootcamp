```python
st1 = states.copy()
```

```python
#ilk 3 sütunumda boş veri olmadığı için geri kalan sütunları dolduruyorum.
for i in st1.columns[3:]:
    for year in st1['YEAR'].unique():
        st1[i].fillna(st1[st1['YEAR'] == year][i].mean(), inplace=True)
```

```python
st1.isnull().sum()
```
