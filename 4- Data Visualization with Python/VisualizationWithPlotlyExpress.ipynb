In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Plotly Express. 

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

**(1)** Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import plotly.express as px
df1 = df[df.location.isin(['Spain', 'France', 'Germany', 'Italy'])]
df2 = df1[df1['date'] >= '2020-04-01']
fig = px.line(df2, x="date", y="total_deaths", color='location', title='Total Deaths of Spain,France,Germany,Italy',
               labels={"date": "Dates",  
                      "total_deaths": "Total Deaths"},
              width = 1200, height = 600)
fig.show()
```

**(2)** Plot a bar plot with plotly express for average death number that compares continents.

```python
avg =df.groupby(by='continent')['total_deaths'].sum()/df.groupby(by='continent')['total_deaths'].count()
avg1 = avg.reset_index()
fig = px.bar(data_frame=avg1,x='continent',y='total_deaths', color='continent')
fig.show()
```

**(3)** Plot a histogram for daily deaths for any country you choose. 

```python
mex = df[df['location']=='Mexico']
fig = px.histogram(mex, x='date', y="new_deaths", title='Daily Deaths of Mexico', 
                   labels={"date": "Date",  
                           "new_deaths": "Daily Deaths"})
fig.show()
```
