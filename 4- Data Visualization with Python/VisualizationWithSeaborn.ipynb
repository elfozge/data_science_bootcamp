In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Seaborn library. 

**(1)** Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

title_style = {'family': 'Arial', 'color': 'darkred', 'size': 20}
axis_style = {'family': 'Arial', 'color': 'darkblue', 'size': 15}

import warnings
warnings.filterwarnings(action="ignore")

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
df1 = df[(df.date >= '2020-04-01') & (df.location.isin(['Spain', 'France', 'Germany', 'Italy']))]
```

```python
plt.figure(figsize=(15, 8), dpi=100)

sns.lineplot(x ='date' , y = 'total_deaths', data = df1, hue = 'location')
plt.title('Total Deaths', fontdict = title_style)
plt.xlabel('Dates', fontdict = axis_style)
plt.ylabel('Count of Deaths', fontdict = axis_style)
plt.show()
```

**(2)** Plot a bar plot with seaborn for average death number that compares continents.

```python
plt.figure(figsize=(15,8))

sns.barplot(x ='continent', y ='total_deaths', data = df)

plt.title('Total Deaths', fontdict = title_style)
plt.xlabel('Continents', fontdict = axis_style)
plt.ylabel('Average Count of Deaths', fontdict = axis_style)

plt.show()
```

**(3)**  Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
df2 = df[df['location'] == 'France']['new_deaths']
```

```python
plt.figure(figsize=(20, 18))

plt.subplot(2,2,1)
sns.distplot(df2.dropna(), bins=50, kde=False)
plt.title('New Deaths\n(Bins:50 / KDE=False)', fontdict=title_style)
plt.xlabel('New Deaths', fontdict=axis_style)
plt.ylabel('Count', fontdict=axis_style)

plt.subplot(2,2,2)
sns.distplot(df2.dropna(), bins=20, kde=False)
plt.title('New Deaths\n(Bins:20 / KDE=False)', fontdict=title_style)
plt.xlabel('New Deaths', fontdict=axis_style)
plt.ylabel('Count', fontdict=axis_style)

plt.subplot(2,2,3)
sns.distplot(df2.dropna(), bins=50, kde=True)
plt.title('New Deaths\n(Bins:50 / KDE=True)', fontdict=title_style)
plt.xlabel('New Deaths', fontdict=axis_style)
plt.ylabel('Normalized Count', fontdict=axis_style)

plt.subplot(2,2,4)
sns.distplot(df2.dropna(), bins=20, kde=True)
plt.title('New Deaths\n(Bins:20 / KDE=True)', fontdict=title_style)
plt.xlabel('New Deaths', fontdict=axis_style)
plt.ylabel('Normalized Count', fontdict=axis_style)

plt.show()
```

**(4)** Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
df3 = df[(df['location'] == 'Italy') | (df['location'] == 'Germany')]
```

```python
plt.figure(figsize=(22, 8))

plt.subplot(131)
sns.boxplot(x = 'location', y = 'new_deaths', data=df3 )
plt.title('Boxplot', fontdict = title_style)
plt.xlabel('Country', fontdict = axis_style)
plt.ylabel('New Deaths', fontdict = axis_style)

plt.subplot(132)
sns.violinplot(x = 'location', y = 'new_deaths', data=df3)
plt.title('Violin Plot', fontdict = title_style)
plt.xlabel('Country', fontdict = axis_style)
plt.ylabel('New Deaths', fontdict = axis_style)


plt.subplot(133)
sns.swarmplot(x = 'location', y = 'new_deaths', data=df3)
plt.title('Swarm Plot', fontdict = title_style)
plt.xlabel('Country', fontdict = axis_style)
plt.ylabel('New Deaths', fontdict = axis_style)

plt.show()
```
