In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)

title_style = {'family': 'Century Gothic', 'color': 'red', 'size': 20 }
axis_style  = {'family': 'Century Gothic', 'color': 'blue', 'size': 15 }
text_style = {'family': 'Century Gothic', 'color': 'black', 'size': 15 }

range_df = df[df["date"].isin(pd.date_range("2020-12-18", periods=100)) &  
              df["location"].isin(['Spain', 'France', 'Germany', 'Italy'])] #son 100 günde İspanya, Fransa, Almanya ve İtalya'nın verilerini aldık

a = range_df[range_df.location == 'France']  
France = a.groupby('date').sum().loc[:,['total_cases','total_deaths']]
b = range_df[range_df.location == 'Spain']
Spain = b.groupby('date').sum().loc[:,['total_cases','total_deaths']]
c = range_df[range_df.location == 'Germany']
Germany = c.groupby('date').sum().loc[:,['total_cases','total_deaths']]
d = range_df[range_df.location == 'Italy']
Italy = d.groupby('date').sum().loc[:,['total_cases','total_deaths']]

# yukarıda ülkelerin sadece toplam vaka ve ölümlü vaka sayılarını alıp tarihlere göre grupladık

plt.figure(figsize=(18,5))

plt.subplot(121)
#plt.text('2020-12-18', 4500000, 'Total Cases', fontdict= text_style) Jupyter notebookta bu satır çalışırken ne yazık ki burada çalısmadı?
plt.xlabel('Date', fontdict = axis_style)
plt.ylabel('Number of Cases', fontdict = axis_style)
plt.plot(France['total_cases'], marker= '*')
plt.plot(Spain['total_cases'], marker = ".")
plt.plot(Germany['total_cases'], lw=2)
plt.plot(Italy['total_cases'],marker = "+")
plt.xticks(rotation = 35, fontsize = 9)
plt.legend(['France','Spain','Germany','Italy'], loc= 4)
plt.title('Coronavirus Statistics', fontdict= title_style, loc='right')

plt.subplot(122)
#plt.text('2020-12-18', 105000, 'Total Deaths', fontdict= text_style) Jupyter notebookta bu satır çalışırken ne yazık ki burada çalısmadı?
plt.xlabel('Date', fontdict = axis_style)
plt.ylabel('Number of Deaths', fontdict = axis_style)
plt.plot(France['total_deaths'], marker= '*')
plt.plot(Spain['total_deaths'], marker = ".")
plt.plot(Germany['total_deaths'], lw=2)
plt.plot(Italy['total_deaths'],marker = "+")
plt.xticks(rotation = 35, fontsize = 9)
plt.legend(['France','Spain','Germany','Italy'], loc=4)


plt.show()
```
