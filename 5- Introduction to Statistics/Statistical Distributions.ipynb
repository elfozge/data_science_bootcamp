```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
distribution=np.random.normal(5, 80, 1000)

plt.figure(figsize=(8,4), dpi = 100)
plt.hist(distribution, bins = 40)
plt.show()
```

<p>In this plot, we have a normal distribution histogram which have 1000 values and also have a mean 5 and standard deviation 80. 95% of the values are above or below two standard deviations of the mean that means they are between -155 and 165. </p>

```python
normally_distribution = np.random.normal(np.mean(distribution),np.std(distribution), 1000)
```

```python
distribution.sort()
normally_distribution.sort()

plt.figure(figsize=(8,5), dpi = 200)
title_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
axis_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
plt.title('Sample Data & Normally Distributed Data', fontdict = title_font)
plt.xlabel('Sample Data', fontdict = axis_font)
plt.ylabel('Normally Distributed Data', fontdict = axis_font)
plt.scatter(normally_distribution, distribution)
plt.plot([np.min(normally_distribution), np.max(normally_distribution)],
         [np.min(distribution), np.max(distribution)], c='red')
plt.show()
```

<p>If the sample data has the same distribution with the normal distribution variable, the result should be a straight line from the bottom left corner to the top right corner. At first sight, Q-Q plot have normal distribution but we have negatively skewed plot. </p>
